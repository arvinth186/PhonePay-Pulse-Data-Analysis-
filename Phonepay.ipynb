```python
# %pip install pymysql

import os
import json
import pandas as pd
import pymysql as mysql
```

```python
# aggre_transactions.py
# here we are reading all the data from the insurance directory 

path1 =  'D:/Guvi Projects/Phonepay/pulse/data/aggregated/transaction/country/india/state/'
agg_trans_list = os.listdir(path1) # list all the files in the directory [this part will use the above string as path]

columns1 = {'States':[],'Years':[],'Quarter':[],'Transaction_type':[],'Transaction_count':[],'Transaction_amount':[]}


for state in agg_trans_list:
    curr_states = path1 + state + '/'
    aggre_years_list = os.listdir(curr_states) # list all the files from the each state directory [2018, 2019, 2020, 2021,...]

    for year in aggre_years_list:
        curr_year = curr_states + year + '/'
        aggre_files_list = os.listdir(curr_year)  # list all the json files from the each year directory [01.json, 02.json, 03.json,...] 

        for file in aggre_files_list:
            curr_file = curr_year + file
            data = open(curr_file,'r') # open the each json file in read mode

            A=json.load(data) # load the json file to read the data
            
            for i in A['data']['transactionData']:
                # as the details in the transaction are in list so we must use the index to access the data
                name=i['name']
                count=i['paymentInstruments'][0]['count']
                amount=i['paymentInstruments'][0]['amount']
                columns1['Transaction_type'].append(name)
                columns1['Transaction_count'].append(count)
                columns1['Transaction_amount'].append(amount)
                columns1['States'].append(state)
                columns1['Years'].append(year)
                columns1['Quarter'].append(int(file.strip('.json'))) # to remove the .json from the file name and convert it into integer which is the quarter number
                
                
```

```python
aggre_transaction = pd.DataFrame(columns1)
aggre_transaction
```

```python
# agrre_user.py
# here we are reading all the data from the user directory


path2 = 'D:/Guvi Projects/Phonepay/pulse/data/aggregated/user/country/india/state/'
agg_user_list = os.listdir(path2)

columns2={'States':[],'Years':[],'Quarter':[],'Brands':[],'Transaction_count':[],'Percentage':[]}

for state in agg_user_list:
    curr_states = path2 + state + '/'
    agg_years_list = os.listdir(curr_states)

    for year in agg_years_list:
        curr_year = curr_states + year + '/'
        aggre_files_list = os.listdir(curr_year)

        for file in aggre_files_list:
            curr_file = curr_year + file
            data = open(curr_file,'r')
            
            B=json.load(data)
            try:
                for i in B['data']['usersByDevice']:
                    brand=i['brand']
                    count=i['count']
                    percentage=i['percentage']
                    columns2['Brands'].append(brand)
                    columns2['Transaction_count'].append(count)
                    columns2['Percentage'].append(percentage)
                    columns2['States'].append(state)
                    columns2['Years'].append(year)
                    columns2['Quarter'].append(int(file.strip('.json')))
            except:
                pass

                


            
```

```python
# map_transaction.py
# here we are reading all the data from the map_transaction directory

path4 = 'D:/Guvi Projects/Phonepay/pulse/data/map/transaction/hover/country/india/state/'
map_insurance_list = os.listdir(path4)

columns4 = {'States':[],'Years':[],'Quarter':[],'Districts':[],'Transaction_count':[],'Transaction_amount':[]}

for state in map_insurance_list:
    curr_states = path4 + state + '/'
    map_years_list = os.listdir(curr_states)

    for year in map_years_list:
        curr_year = curr_states + year + '/'
        map_files_list = os.listdir(curr_year)

        for file in map_files_list:
            curr_file = curr_year + file
            data = open(curr_file,'r')

            D=json.load(data)
            # when we try to read metrics data it appear as list so we have to use index[0] to access the data
            
            for i in D['data']['hoverDataList']:
                name=i['name']
                count=i['metric'][0]['count']
                amount=i['metric'][0]['amount']
                columns4['Districts'].append(name)
                columns4['Transaction_count'].append(count)
                columns4['Transaction_amount'].append(amount)
                columns4['States'].append(state)
                columns4['Years'].append(year)
                columns4['Quarter'].append(int(file.strip('.json')))

                
```

```python
# map_user.py
# here we are reading all the data from the map_user directory

path5 = 'D:/Guvi Projects/Phonepay/pulse/data/map/user/hover/country/india/state/'
map_user_list = os.listdir(path5)

columns5 = {'States':[],'Years':[],'Quarter':[],'Districts':[],'RegisteredUsers':[],'AppOpens':[]}

for state in map_user_list:
    curr_states = path5 + state + '/'
    map_years_list = os.listdir(curr_states)

    for year in map_years_list:
        curr_year = curr_states + year + '/'
        map_files_list = os.listdir(curr_year)

        for file in map_files_list:
            curr_file = curr_year + file
            data = open(curr_file,'r')

            E = json.load(data)

            for i in E['data']['hoverData'].items():
                district=i[0]
                registeredUsers=i[1]['registeredUsers']
                appOpens=i[1]['appOpens']
                columns5['Districts'].append(district)
                columns5['RegisteredUsers'].append(registeredUsers)
                columns5['AppOpens'].append(appOpens)
                columns5['States'].append(state)
                columns5['Years'].append(year)
                columns5['Quarter'].append(int(file.strip('.json')))
```

```python
conn = mysql.connect(
    host='127.0.0.1',
    user='root',
    password='root',
    database='phonepay'
)

cursor = conn.cursor()
cursor.execute("select database();")
print(cursor.fetchone()) # as it the 1st line of the result from mysql

# aggre_insurance table creation in mysql
create_table_agg_insurance = '''create table if not exists aggregated_insurance (States varchar(300),
                                                                    Years int,
                                                                    Quarter int,
                                                                    Name varchar(300),
                                                                    Insurance_count bigint,
                                                                    TotalIns_amount bigint);'''
cursor.execute(create_table_agg_insurance)
conn.commit()
print("Aggregated Insurance table created successfully in MySQL")

insert_agg_insurance = '''insert into aggregated_insurance (States, Years, Quarter, Name, Insurance_count, TotalIns_amount)
                          values (%s, %s, %s, %s, %s, %s)'''

data = aggre_insurance.values.tolist()
cursor.executemany(insert_agg_insurance,data)
conn.commit()
print(cursor.rowcount, "Records inserted successfully into aggregated_insurance table")

# aggre_transaction table creation in mysql
create_table_agg_transaction = '''create table if not exists aggregated_transaction (States varchar(300),
                                                                    Years int,
                                                                    Quarter int,
                                                                    Transaction_type varchar(300),
                                                                    Transaction_count bigint,
                                                                    Transaction_amount bigint);'''
cursor.execute(create_table_agg_transaction)
conn.commit()
print("Aggregated Transaction table created successfully in MySQL")

insert_agg_transaction = '''insert into aggregated_transaction (States, Years, Quarter, Transaction_type, Transaction_count, Transaction_amount)
                          values (%s, %s, %s, %s, %s, %s)'''

data = aggre_transaction.values.tolist() 
cursor.executemany(insert_agg_transaction,data) 
conn.commit()
print(cursor.rowcount, "Records inserted successfully into aggregated_transaction table")

# aggre_user table creation in mysql
create_table_agg_user = '''create table if not exists aggregated_user (States varchar(300),
                                                                    Years int,
                                                                    Quarter int,
                                                                    Brands varchar(300),
                                                                    Transaction_count bigint,
                                                                    Percentage float);'''

cursor.execute(create_table_agg_user)
conn.commit()
print("Aggregated User table created successfully in MySQL")

insert_agg_user = '''insert into aggregated_user (States, Years, Quarter, Brands, Transaction_count, Percentage)
                          values (%s, %s, %s, %s, %s, %s)'''
data = aggre_user.values.tolist()
cursor.executemany(insert_agg_user,data)
conn.commit()
print(cursor.rowcount, "Records inserted successfully into aggregated_user table")

# map_insurance table creation in mysql
create_table_map_insurance = '''create table if not exists map_insurance (States varchar(300),
                                                                    Years int,
                                                                    Quarter int,
                                                                    Districts varchar(300),
                                                                    Insurance_count bigint,
                                                                    TotalIns_amount bigint);'''

cursor.execute(create_table_map_insurance)
conn.commit()
print("Map Insurance table created successfully in MySQL")

insert_map_insurance = '''insert into map_insurance (States, Years, Quarter, Districts, Insurance_count, TotalIns_amount)
                          values (%s, %s, %s, %s, %s, %s)'''

data = map_insurance.values.tolist()
cursor.executemany(insert_map_insurance,data)
conn.commit()
print(cursor.rowcount, "Records inserted successfully into map_insurance table")


# map_transaction table creation in mysql
create_table_map_transaction = '''create table if not exists map_transaction (States varchar(300),
                                                                    Years int,
                                                                    Quarter int,
                                                                    Districts varchar(300),
                                                                    Transaction_count bigint,
                                                                    Transaction_amount bigint);'''      
cursor.execute(create_table_map_transaction)
conn.commit()
print("Map Transaction table created successfully in MySQL")

insert_map_transaction = '''insert into map_transaction (States, Years, Quarter, Districts, Transaction_count, Transaction_amount)
                            values (%s, %s, %s, %s, %s, %s)'''  

data = map_transaction.values.tolist()
cursor.executemany(insert_map_transaction,data)
conn.commit()
print(cursor.rowcount, "Records inserted successfully into map_transaction table")

# map_user table creation in mysql
create_table_map_user = '''create table if not exists map_user (States varchar(300),
                                                                    Years int,
                                                                    Quarter int,
                                                                    Districts varchar(300),
                                                                    RegisteredUsers bigint,
                                                                    AppOpens bigint);'''
cursor.execute(create_table_map_user)
conn.commit()
print("Map User table created successfully in MySQL")

insert_map_user = '''insert into map_user (States, Years, Quarter, Districts, RegisteredUsers, AppOpens)
                        values (%s, %s, %s, %s, %s, %s)'''

data = map_user.values.tolist()
cursor.executemany(insert_map_user,data)
conn.commit()
print(cursor.rowcount, "Records inserted successfully into map_user table")

# top_insurance table creation in mysql

create_table_top_insurance = '''create table if not exists top_insurance (States varchar(300),
                                                                    Years int,
                                                                    Quarter int,
                                                                    Pincodes int,
                                                                    Insurance_count bigint,
                                                                    TotalIns_amount bigint);'''
cursor.execute(create_table_top_insurance)
conn.commit()
print("Top Insurance table created successfully in MySQL")

insert_top_insurance = '''insert into top_insurance (States, Years, Quarter, Pincodes, Insurance_count, TotalIns_amount)
                          values (%s, %s, %s, %s, %s, %s)'''

data = top_insurance.values.tolist()
cursor.executemany(insert_top_insurance,data)
conn.commit()
print(cursor.rowcount, "Records inserted successfully into top_insurance table")    

# top_transaction table creation in mysql
create_table_top_transaction = '''create table if not exists top_transaction (States varchar(300),
                                                                    Years int,
                                                                    Quarter int,
                                                                    Pincodes int,
                                                                    Transaction_count bigint,
                                                                    Transaction_amount bigint);'''
cursor.execute(create_table_top_transaction)
conn.commit()
print("Top Transaction table created successfully in MySQL")

insert_top_transaction = '''insert into top_transaction (States, Years, Quarter, Pincodes, Transaction_count, Transaction_amount)
                            values (%s, %s, %s, %s, %s, %s)'''

data = top_transaction.values.tolist()
cursor.executemany(insert_top_transaction,data)
conn.commit()
print(cursor.rowcount, "Records inserted successfully into top_transaction table")  

# top_user table creation in mysql
create_table_top_user = '''create table if not exists top_user (States varchar(300),
                                                                    Years int,
                                                                    Quarter int,
                                                                    Pincodes int,
                                                                    RegisteredUsers bigint);'''
cursor.execute(create_table_top_user)
conn.commit()
print("Top User table created successfully in MySQL")

insert_top_user = '''insert into top_user (States, Years, Quarter, Pincodes, RegisteredUsers)
                     values (%s, %s, %s, %s, %s)'''

data = top_user.values.tolist()
cursor.executemany(insert_top_user,data)
conn.commit()
print(cursor.rowcount, "Records inserted successfully into top_user table")

```
